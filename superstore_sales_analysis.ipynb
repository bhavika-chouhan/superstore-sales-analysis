```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'sub-category' : 'sub_category'})
```

```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
category_sales = df.groupby('category')['sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
category_profit = df.groupby('category')['profit'].sum().sort_values(ascending=False)
print(category_profit)
```

```python
category_summary = pd.DataFrame({
    'sales' : category_sales,
    'profit' : category_profit 
    })
print(category_summary)
```

```python
category_summary.plot(kind='bar', color=['skyblue','orange'], figsize=(5,4))
plt.title("Total Sales and Profit by Category")
plt.ylabel("Amount")
plt.xlabel("")
plt.xticks(rotation=0)
plt.legend(loc='upper left', fontsize=8)
```

```python
sub_category_profit = df.groupby('sub_category')['profit'].sum().sort_values(ascending=True).head(13)
print(sub_category_profit)
```

```python
sub_category_sales = df.groupby('sub_category')['sales'].sum().sort_values(ascending=True).head(10)
print(sub_category_sales)
```

```python
sub_category_sales.plot(kind='barh', color='skyblue')
plt.title("Top 10 Sub-Categories by Sales")
plt.ylabel("")
plt.xlabel("Total Sales")
plt.show()
```

```python
color = ['red' if val < 0 else 'green' for val in sub_category_profit ]
sub_category_profit.plot(kind='barh',color=color)
plt.title("Sub-Categories Profit (Profit vs Loss)")
plt.ylabel("")
plt.xlabel("Total Profit & Loss")
plt.axvline(0,linewidth=1,color='black')
```

```python
monthly_sales = df.groupby('order_month')['sales'].sum().sort_index()
print(monthly_sales)
```

```python
monthly_sales.index = monthly_sales.index.astype(str)
```

```python
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Months')
plt.ylabel('Sales')
```

```python
monthly_profit = df.groupby('order_month')['profit'].sum().sort_index()
print(monthly_profit)
```

```python
monthly_profit.index = monthly_profit.index.astype(str)
```

```python
plt.plot(monthly_profit.index, monthly_profit.values, marker='o')
plt.title('Monthly Profit Trend')
plt.xlabel('Months')
plt.ylabel('Profit')
```
